```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd 
```

```python
url = 'http://www.nuuo.com/Nation.php?WC_ID='
regions = ['0001', '0002', '0003', '0004', '0005', '0006']
```

```python
r = urllib.request.urlopen((url + regions[0])).read()
soup = BeautifulSoup(r, 'html.parser')
print(soup)
```

```python
soup.find_all('a', class_='type_url01')
```

```python
links = []

for region in regions: 
    r = urllib.request.urlopen((url + region)).read()
    link_soup = BeautifulSoup(r, 'html.parser')
    for link in link_soup.find_all('a', 'type_url01'):
        links.append(link)
        print(link)
```

```python
parsed_links = [(href.get_text(), href['href']) for href in links]
```

```python
parsed_links
```

```python
core_url = 'http://www.nuuo.com/'

link = links[0]
r = urllib.request.urlopen(core_url+link[1]).read()
country_soup = BeautifulSoup(r, 'html.parser')
# for country in links: 
#     print(core_url+country[1])
```

```python
companies = []

for div in country_soup.find_all('div', class_='partner-item'):
    company = [td.get_text().strip() for td in div.find_all('td')]
#     companies.append(div_soup)
    co = {}
    co['name'] = company[1]
    print(company)
    try:
        co['phone'] = company[company.index('Phone:')+1]
        co['address'] = company[company.index('Address:')+1]
        co['email'] = [x for x in company if 'Email:' in x][0][7:]
    except:
        pass
        
    companies.append(co)
```

```python
all_companies = []

core_url = 'http://www.axxonsoft.com'

# link = links[0]
for country_tuple in links: 
    country = country_tuple[0]
    r = urllib.request.urlopen(core_url+country_tuple[1]).read()
    country_soup = BeautifulSoup(r, 'html.parser')
    for div in country_soup.find_all('div', class_='partner-item'):
        company = [td.get_text().strip() for td in div.find_all('td')]
    #     companies.append(div_soup)
        co = {}
        co['name'] = company[1]
        try:
            co['phone'] = company[company.index('Phone:')+1]
            co['address'] = company[company.index('Address:')+1]
            co['email'] = [x for x in company if 'Email:' in x][0][7:]
            co['country'] = country
        except:
            pass

        all_companies.append(co)
```
