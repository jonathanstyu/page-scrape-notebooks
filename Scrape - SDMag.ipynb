```python
import pandas as pd
import numpy as np
import urllib.request
from bs4 import BeautifulSoup
from bs4 import Comment
import time 
```

```python
letters = ['A', 'C', 'D', 'E', "F", 'H', 'I', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'U', 'W', 'Z']
url = "http://www.sdmmag.com/directories/2743-central-stations-monitoring-guide/{0}"
lettered_urls = [url.format(x) for x in letters]
```

```python
url_base = 'http://www.sdmmag.com/directories/2169?id_raw=2169-buyers-guide&page={0}'
urls = [url_base.format(x) for x in range(1, 16)]
urls
```

```python
soups = []
articles = []
for url in urls: 
#     print(url)
    r = urllib.request.urlopen(url)
    soups.append(BeautifulSoup(r, 'html.parser'))
```

```python
for soup in soups: 
    articles += soup.find_all('article', class_='directory-listing')
```

```python
links = [article.find('a')['href'] for article in articles]
formatted = ['http://www.sdmmag.com/' + link for link in links]
```

```python
# len(pro_soups)
pro_soups[45]
```

```python
profiles = []
pro_soups = []

for z in formatted: 
#     print(z)
#     time.sleep(5)
    r = urllib.request.urlopen(z)
    pro_soups.append(BeautifulSoup(r, 'html.parser'))
```

```python
profiles += [soup.find_all('div', class_='directory-listing-profile__details') for soup in pro_soups]
# profs = [thing.find('div', class_='') for thing in thingy]
```

```python
flattened = [profile[0] for profile in profiles]
```

```python
len(flattened)
```

```python
cos= []

for prof in flattened: 
# prof = profs[27] 
    co = {
        'name': prof.find('div', 'directory-listing-profile__company-name').get_text().strip() if prof.find('div', 'directory-listing-profile__company-name') else '',
        'phone': prof.find('div', 'directory-listing-profile__phone').get_text().strip().split(':')[1] if prof.find('div', 'directory-listing-profile__phone') else '',
        'country': prof.find('div', 'directory-listing-profile__country').get_text().strip() if prof.find('div', 'directory-listing-profile__country') else '',
        'url': prof.find('div', 'directory-listing-profile__url-link').get_text().strip() if prof.find('div', 'directory-listing-profile__url-link') else '',
        'email': prof.find('div', 'directory-listing-profile__email').get_text().strip() if prof.find('div', 'directory-listing-profile__email') else '',
        'address1': prof.find('div', 'directory-listing-profile__address1').get_text().strip() if prof.find('div', 'directory-listing-profile__address1') else '',
        'city': prof.find('div', 'directory-listing-profile__city').get_text().split('\n')[1].strip() if prof.find('div', 'directory-listing-profile__city') else '',
        'state': prof.find('div', 'directory-listing-profile__city').get_text().split('\n')[2].strip() if prof.find('div', 'directory-listing-profile__city') else '',
        'zip': prof.find('div', 'directory-listing-profile__city').get_text().split('\n')[3].strip() if prof.find('div', 'directory-listing-profile__city') else ''
    }

    for x, div in enumerate(prof.find_all('div', 'directory-listing-custom-contacts__contact')):
        co['contact{0}-name'.format(x)] = div.find('div', class_='directory-listing-custom-contacts__contact-name').get_text().strip() if div.find('div', class_='directory-listing-custom-contacts__contact-name') else ''
        co['contact{0}-email'.format(x)] = div.find('div', class_='directory-listing-custom-contacts__contact-email').get_text() if div.find('div', class_='directory-listing-custom-contacts__contact-email') else ''
        co['contact{0}-phone'.format(x)] = div.find('div', class_='directory-listing-custom-contacts__contact-phone').get_text() if div.find('div', class_='directory-listing-custom-contacts__contact-phone') else ''

    cos.append(co)
```

```python
cos = cos[:35] + cos[36:]
cos[35]
```
