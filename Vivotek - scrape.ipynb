```python
from bs4 import BeautifulSoup
import urllib.request

url = 'http://www.axxonsoft.com/partners/partner_locator/wheretobuy/'

r = urllib.request.urlopen(url).read()
soup = BeautifulSoup(r, 'html.parser')
print(soup)
```

```python
list_boxes = soup.find_all('div', class_='list-item box')
```

```python
box = list_boxes[0]
box
```

```python
span = box.find('li', 'jplist_wheretobuy_des_tel').find('a').get_text()
print(span)
```

```python
buyers = []

for box in list_boxes:
    buy = {}
    try:
        buy['address'] = box.find('address').get_text()
        buy['title'] = box.find('h1', 'jplist_wheretobuy_title').get_text().strip()
        buy['country'] = box.find('span', 'jplist_wheretobuy_country').get_text().strip()
        buy['phone'] = box.find('li', 'jplist_wheretobuy_des_tel').find('a').get_text()
        buy['mail'] = box.find('li', 'jplist_wheretobuy_des_mail').find('a').get_text()
    except:
        print("UHOH")
    
    buyers.append(buy)
```

```python
import pandas as pd

pd.DataFrame(buyers).to_csv('pandas')
```
