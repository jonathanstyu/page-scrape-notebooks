```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd 
import csv
```

Hikvision was not as challenging when it came to scraping leads. They had a number of different web pages broken down by country. The HTML is not labeled individually by the piece of info, but at the same time it is clean and well structured. 

Previously I was able to scrape the raw HTML out of the page, accumulated as a single CSV. 

```python
htmlfiles = []

# pre downloaded because I am lazy

with open('Hikivision - Sheet1.csv', 'rt') as csvfile: 
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader: 
        htmlfiles.append(row[0])
```

```python
soups = [BeautifulSoup(x, 'html.parser') for x in htmlfiles]
```

```python
soups[0]
```

```python
companies = []
for soup in soups: 
    company = {}
    links = soup.find('div', class_='main_right').find_all('a')
    company['name'] = soup.find('div', class_='main_right').find_all('p')[0].get_text()
    company['email'] = links[0].get_text()
    company['website'] = links[1].get_text()
    companies.append(company)
```

```python
# Convert to a DataFrame

comp_leads = pd.DataFrame(companies)
comp_leads
```

```python
# Export out to a csv

comp_leads.to_csv('hikvision.csv')
```
