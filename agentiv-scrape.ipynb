```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = 'https://www.agentvi.com/how-to-buy'

user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,} 
# r = urllib.request.urlopen(url).read()
request=urllib.request.Request(url,None,headers) #The assembled request
response = urllib.request.urlopen(request)
data = response.read() # The data u need
soup = BeautifulSoup(data, 'html.parser')
print(soup)
```

```python
links = []

for country in soup.find_all('option'):
    if country["value"]=="":
        pass
    else: 
        try: 
            links.append(country['value'])
        except:
            pass
```

```python
country_link = 'https://www.agentvi.com/how-to-buy/{0}'
soups = []

user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,} 

for country in links: 
    request=urllib.request.Request(country_link.format(country),None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
    soups.append(BeautifulSoup(data, 'html.parser'))
```

```python
ints = []

for soup in soups:
    for link in soup.find_all('div', class_="ls-certified-systems-integrator"):
        ints.append({
            "url": link.find('a')['href'],
            "name": link.find('img')['title'],
            'country': soup.find("h1").get_text(),
            'companyType': "System Integrator"
        })
    for link in soup.find_all('div', class_="ls-distributor"):
        ints.append({
            "url": link.find('a')['href'],
            "name": link.find('img')['title'],
            'country': soup.find("h1").get_text(),
            'companyType': "Distributor"
        })

print(ints)
```

```python
import pandas as pd

agentiv = pd.DataFrame(ints)
agentiv.to_csv('agentiv.csv')
```
