```python
from bs4 import BeautifulSoup

# I had scraped the ONVIF list earlier https://www.onvif.org/about/member-list/
import onvif_output
```

```python
rawl_html = onvif_output.items
```

```python
parser = BeautifulSoup(rawl_html[2]['Field1'], 'html.parser')
```

```python
# Played around with the settings so that I could see which goes where

parser.find('td', class_='ng-binding')
```

```python
companies = []

for item in rawl_html[2:]: 
    parser = BeautifulSoup(item['Field1'], 'html.parser')
    try: 
        companies.append({
            'name': parser.find('td', class_='ng-binding').get_text(),
            'image_file': parser.find('img')['src'],
            'website': parser.find('a')['href']
        })
    except: 
        pass
```

```python
# Exported objects

companies
```

```python
import pandas as pd 

# put it into a csv

comp = pd.DataFrame(companies)
comp.to_csv('competitors.csv')
```
